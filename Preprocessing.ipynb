## Setting Up API

````python
import os
import googleapiclient.discovery

# Set up the YouTube API client
api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = "Your API"

youtube = googleapiclient.discovery.build(api_service_name, api_version, developerKey=DEVELOPER_KEY)
````

- `query:` The search query string.

- `max_results` (optional): The maximum number of results to return, with a default value of 100.

- it creates a request object using `youtube.search().list():`

- The `request.execute()` method is called to execute the search request and retrieve the response.

- The `request.execute()` method is called to execute the search request and retrieve the response.

````python
def search_videos(query, max_results=100):
    request = youtube.search().list(
      q=query,
      part='snippet',
      type='video',
      videoCaption='closedCaption',
      relevanceLanguage='en',
      maxResults=max_results
    )
    response = request.execute()
    return response.get('items', [])
````

`collect_video_data` is to gather video data based on different sentiment categories (positive, negative, neutral). It uses a search function ( `search_videos`) to retrieve video information from a video platform API (like YouTube).

### Structure

1. **Queries Dictionary**: It defines three lists of search queries based on sentiment.

2. **Video Data Storage**: It initializes a dictionary to store video data categorized by sentiment.

3. **Search Loop**: It iterates through each sentiment category and its associated queries, performing a search for each query and collecting relevant video data.

### Key Components
- **`queries`**: A dictionary mapping sentiment categories to search queries.
- **`video_data`**: A dictionary initialized to hold lists of video data for each sentiment.
- **Nested Loops**: The outer loop iterates through sentiments, and the inner loop processes each search query.
- **Data Collection**: For each video result, it extracts the title, description, and video ID, appending this information to the appropriate list in `video_data`.

````python
def collect_video_data():

    queries = {
        'positive': ['motivational speech english', 'success story', 'inspirational talk','Positive message'],
        'negative': ['negative news', 'angry rant', 'critical review', 'gun voilence'],
        'neutral': ['tutorial', 'product description', 'informational video']
    }

    video_data = {'positive': [], 'negative': [], 'neutral': []}

    for sentiment, q_list in queries.items():
        for query in q_list:
            results = search_videos(query)
            for item in results:
                video_data[sentiment].append({
                    'title': item['snippet']['title'],
                    'description': item['snippet']['description'],
                    'videoId': item['id']['videoId']
                })

    return video_data

video_data = collect_video_data()
````

````python
# Print collected video data
for sentiment in video_data:
    print(f"\n{sentiment.capitalize()} videos:")
    for video in video_data[sentiment]:
        print(f"Title: {video['title']}, Video ID: {video['videoId']}")
````

## Doing all the nessesary preprocessing

````python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def filter_text(text):
  if isinstance(text, str): # Check if the text is a string
    text = text.lower() # Convert to lowercase

    # Removing punctuation
    extra_chars = string.punctuation + '\t\n'
    text= text.translate(str.maketrans('', '', extra_chars))

    # Removing extra spaces
    text = re.sub(r'\s+', ' ', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # Remove special characters and numbers
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'\d', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]

    return ' '.join(tokens)
  else:
      print(f"Non-string value encountered: {text}")
      return "" # Return an empty string if text is not a string
````
